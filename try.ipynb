```python
import matplotlib.pyplot as plt

# Data from your logs
epochs = [1, 2, 3, 4, 5, 6]
train_loss = [25.6620, 7.0946, 4.4299, 3.3783, 2.7862, 2.4083]
val_loss = [19.9441, 7.5967, 5.0078, 2.7365, 5.9249, 5.1181]

# Plot the losses
plt.plot(epochs, train_loss, label="Train Loss", color="tab:blue")
plt.plot(epochs, val_loss, label="Validation Loss", color="tab:orange")

# Labels and title
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")

# Legend
plt.legend()

# Show the plot
plt.show()
```
