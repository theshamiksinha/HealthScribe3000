```python
import matplotlib.pyplot as plt

# Data
epochs = [1, 2, 3, 4, 5, 6]
em_scores = [28.04, 40.44, 44.65, 46.96, 45.42, 46.25]  # Updated EM scores

# Plot
plt.figure(figsize=(8, 6))
plt.plot(epochs, em_scores, marker='o', linestyle='-', markersize=6, label="Exact Match (EM) Score", color='C0')

# Labels and title
plt.xlabel("Epoch")
plt.ylabel("Exact Match Score (%)")
plt.title("Exact Match Score Over Epochs")

# Remove background grid
plt.grid(False)

# Legend
plt.legend()

# Show plot
plt.show()
```
